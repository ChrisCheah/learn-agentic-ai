```python
# ============================================================
# INSTALL REQUIRED LIBRARY
# ============================================================
%pip install openai  python-dotenv --quiet

# ============================================================
# IMPORTS
# ============================================================
from openai import OpenAI
import os
from dotenv import load_dotenv

# Load .env file from the current directory
load_dotenv()

# ============================================================
# ENTER YOUR API KEY HERE
# ============================================================
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
# ============================================================
# INITIALIZE CLIENT
# ============================================================
client = OpenAI(api_key=OPENAI_API_KEY)

# ============================================================
# SIMPLE CHAT FUNCTION (FIXED VERSION)
# ============================================================
def chat(message: str) -> str:
    """
    Send a message to GPT-4o and return the model's reply.
    This version uses the correct attribute access for the new SDK.
    """
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "user", "content": message}
            ]
        )

        # ------------------------------------------------------------
        # FIXED: The new SDK stores message content in:
        # response.choices[0].message.content
        # ------------------------------------------------------------
        return response.choices[0].message.content

    except Exception as e:
        raise RuntimeError(f"Chat request failed: {e}")

# ============================================================
# TEST CONNECTION
# ============================================================
print(chat("Hello! Testing the fixed version."))
```
